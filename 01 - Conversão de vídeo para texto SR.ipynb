## 2o.A - Transcrição do áudio no idioma original usando bibliotecas tradicionais

```python
# Dividindo o arquivo .WAV em pedaços menores
tamanho = 30000   # 30 segundos

partes1 = make_chunks (audio1, tamanho)
partes_audio1 =[]
for i, parte1 in tqdm(enumerate(partes1)):
    parte_name1 = 'audio_1_{0}.wav'.format(i)
    partes_audio1.append(parte_name1)
    parte1.export(parte_name1, format='wav')

partes2 = make_chunks (audio2, tamanho)
partes_audio2 =[]
for i, parte2 in tqdm(enumerate(partes2)):
    parte_name2 = 'audio_2_{0}.wav'.format(i)
    partes_audio2.append(parte_name2)
    parte2.export(parte_name2, format='wav')

partes3 = make_chunks (audio3, tamanho)
partes_audio3 =[]
for i, parte3 in tqdm(enumerate(partes3)):
    parte_name3 = 'audio_3_{0}.wav'.format(i)
    partes_audio3.append(parte_name3)
    parte3.export(parte_name3, format='wav')

partes4 = make_chunks (audio4, tamanho)
partes_audio4 =[]
for i, parte4 in tqdm(enumerate(partes4)):
    parte_name4 = 'audio_4_{0}.wav'.format(i)
    partes_audio4.append(parte_name4)
    parte4.export(parte_name4, format='wav')

partes5 = make_chunks (audio5, tamanho)
partes_audio5 =[]
for i, parte5 in tqdm(enumerate(partes5)):
    parte_name5 = 'audio_5_{0}.wav'.format(i)
    partes_audio5.append(parte_name5)
    parte5.export(parte_name5, format='wav')

partes6 = make_chunks (audio6, tamanho)
partes_audio6 =[]
for i, parte6 in tqdm(enumerate(partes6)):
    parte_name6 = 'audio_6_{0}.wav'.format(i)
    partes_audio6.append(parte_name6)
    parte6.export(parte_name6, format='wav')
```

```python
#  Instalação da biblioteca para Reconhecimento de fala, com suporte para diversos engines e APIs, online e offline.
!pip install SpeechRecognition

# Importando as biblioteca
import speech_recognition as sr   # Reconhecimento de fala, com suporte para diversos engines e APIs, online e offline.
```

```python
# Realizando a transcrição do audio em texto usando Google
def transcreve_audio(nome_audio):
  r = sr.Recognizer()
  with sr.AudioFile(nome_audio) as source:
    audio = r.record(source)
  try:
    texto = r.recognize_google(audio,language='ja-JP')
    print('Google Speech Recognition: ' + texto)
  except sr.UnknownValueError:
    texto = ''
    print('Google Speech Recognition NÃO ENTENDEU o audio')
  except sr.RequestError as e:
    texto = ''
    print('Erro ao solicitar resultados do Google Speech Recognition; {0}'.format(e))
  #finally:
   # try:
   #     os.remove(nome_audio)
   # except:
   #     print('Arquivo não existe')
  return texto
```

```python
# Armazena a conversão do texto completo
texto_1 = []
for parte in tqdm(partes_audio1):
  #texto = texto + ' ' + transcreve_audio(parte)  #aqui esta gravando o texto todo em uma unica linha
  texto_1.append([parte,transcreve_audio(parte)])
# breakc
# preciso gravar em dataframe cada linha convertida para depois fazer a tradução por pedaços
arquivo_Transcricao_1 = pd.DataFrame(texto_1,columns=['parte','transcricao'])
arquivo_Transcricao_1.to_csv(f'{arquivo1}.csv')
```
