```python
#  Avaliar os textos gerados da transcrição do vídeo versus o texto original

# Texto original - tradução oficial - transcrições das legendas oficiais
arq_10 = open("Texto10.txt")
linhas_arq_10 = arq_10.readlines()
for linha_arq_10 in linhas_arq_10:
  linha_arq_10

# Texto gerado pelo arquivo de vídeo - traduzido pela biblioteca
arq_1 = open("Texto1.txt")
linhas_arq_1 = arq_1.readlines()
for linha_arq_1 in linhas_arq_1:
  linha_arq_1

#Sentences are encoded by calling model.encode()
emb10 = model.encode(linhas_arq_10)
emb1  = model.encode(linhas_arq_1)

cos_sim_10 = util.cos_sim(emb10, emb1)
print("Cosine-Similarity by Meu Primeiro Amigo é um Tapado - S1E1:", cos_sim_10)
```

```python
#  Avaliar os textos gerados da transcrição do vídeo versus o texto original

# Texto original - tradução oficial
arq_11 = open("Texto11.txt")
linhas_arq_11 = arq_11.readlines()
for linha_arq_11 in linhas_arq_11:
  linha_arq_11

# Texto gerado pelo arquivo de vídeo completo
arq_2 = open("Texto2.txt")
linhas_arq_2 = arq_2.readlines()
for linha_arq_2 in linhas_arq_2:
  linha_arq_2

#Sentences are encoded by calling model.encode()
emb11 = model.encode(linhas_arq_11)
emb2  = model.encode(linhas_arq_2)

cos_sim_11 = util.cos_sim(emb11, emb2)
print("Cosine-Similarity by Those Snow White Notes - S1E2:", cos_sim_11)
```

```python
# Avaliar os textos gerados da transcrição do vídeo musical e vocal versus o texto original ----- MUSICA - Fuyu No Hanashi - Centimillimental

# Texto original - transcrição da letra da musica em português - tradução oficial
arq_12 = open("Texto12.txt")
linhas_arq_12 = arq_12.readlines()
for linha_arq_12 in linhas_arq_12:
  linha_arq_12

# Texto gerado pelo arquivo de vídeo completo
arq_3 = open("Texto3.txt")
linhas_arq_3 = arq_3.readlines()
for linha_arq_3 in linhas_arq_3:
  linha_arq_3

# Texto gerado pelo arquivo de audio contendo apenas o vocal
arq_5 = open("Texto5.txt")
linhas_arq_5 = arq_5.readlines()
for linha_arq_5 in linhas_arq_5:
  linha_arq_5

#Sentences are encoded by calling model.encode()
emb12 = model.encode(linhas_arq_12,convert_to_tensor=True)
emb3  = model.encode(linhas_arq_3,convert_to_tensor=True)
emb5  = model.encode(linhas_arq_5,convert_to_tensor=True)

cos_sim_12 = util.cos_sim(emb12, emb3)
cos_sim_3  = util.cos_sim(emb12, emb5)
cos_sim_5  = util.cos_sim(emb3 , emb5)

print("Cosine-Similarity by Fuyu No Hanashi - Centimillimental Original versus Arq3:", cos_sim_12)
print("Cosine-Similarity by Fuyu No Hanashi - Centimillimental Original versus Arq5:", cos_sim_3)
print("Cosine-Similarity by Fuyu No Hanashi - Centimillimental Arq3 versus Arq5:"    , cos_sim_5)
```

```python
import spacy
# Carregando o modelo de linguagem (neste caso large)
!python -m spacy download pt_core_news_lg

# Aplicando o modelo de liguagem
nlp = spacy.load('pt_core_news_lg')

# originais
doc10 = nlp(linha_arq_10)
doc11 = nlp(linha_arq_11)
doc12 = nlp(linha_arq_12)
doc13 = nlp(linha_arq_13)

# transcritos
doc1 = nlp(linha_arq_1)
doc2 = nlp(linha_arq_2)
doc3 = nlp(linha_arq_3)
doc4 = nlp(linha_arq_4)
doc5 = nlp(linha_arq_5)
doc6 = nlp(linha_arq_6)
```

```python
# pegar todos os tokens em forma de string para todos os tokens dentro do documento

tokens_10 = [token.orth_ for token in doc10]
tokens_11 = [token.orth_ for token in doc11]
tokens_12 = [token.orth_ for token in doc12]
tokens_13 = [token.orth_ for token in doc13]

tokens_1 = [token.orth_ for token in doc1]
tokens_2 = [token.orth_ for token in doc2]
tokens_3 = [token.orth_ for token in doc3]
tokens_4 = [token.orth_ for token in doc4]
tokens_5 = [token.orth_ for token in doc5]
tokens_6 = [token.orth_ for token in doc6]
```
